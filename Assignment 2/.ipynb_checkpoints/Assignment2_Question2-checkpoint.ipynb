# Question 2_Part1

1. Read all the json files in the folder called Data.
2. There are three categories of json files in this folder. They are identified by the key called “term” in each of the json file.
3. Create a folder structure to read all these json files and store them into these separate folders. You are expected to create a hierarchy of folder structure.
4. Example:
5. You can place all restaurants json files in a particular country (say Australia) in the same folder. How you group the json files and create a folder structure is your choice. Your task is to identify criteria by which you can group all these json files and store them.
6. (You could use these keys to create hierarchy and store json files: Country, city, categories)

* Output Format:
1. Create a folder (Name: Data Processed)
2. In this folder you should have a hierarchy of folder structures (Example: Data Processed/Australia(AU)/........)
3. A good idea is that you can classify json files on Country name or code (You can create a hierarchy of folder structures to effectively sort and store the files). The original json files in the folder “Data” have the name of the “id” key in the file. You can even think of changing the name of the json file when you read and store them.
4. Submission: You will be submitting the Folder (Data Processed )which is storing the json files and the jupyter notebook (code) that you used to create this hierarchy of folder structures.

```python
# import os, json and shutil
import os, json
import shutil
```

```python
# get all the json files in the Data folder
path_to_json='Data/'
json_files = [jsonfile for jsonfile in os.listdir(path_to_json) if jsonfile.endswith('.json')]

# open and read each json file in the Data folder.
for js in json_files:
    with open(os.path.join(path_to_json, js)) as json_file:
        #convert JSON to python List Object
        alldata= json.load(json_file)
        
        #get the data of the category        
        category = alldata['term']
        
        #get the data of the country       
        country = alldata['location']['country']
        
        #get the data of the state        
        state = alldata['location']['state']
        
        #get the data of the city                
        city = alldata['location']['city']
        
        #create new folder path according to their category,country,state and city
        newpath_c = 'Data Processed/'+ category +'/'+country+'/'+ state +'/'+ city
        
        #if the folder path is new, then create a new folder and copy the json file from Data to new folder
        if not os.path.exists(newpath_c):
            os.makedirs(newpath_c)
            shutil.copy('Data/'+js ,'Data Processed/'+ category +'/'+country +'/'+ state +'/'+ city+'/'+ js)
        #if the folder is already exist,then only copy the json without creating a new folder.
        else:
            shutil.copy('Data/'+js ,'Data Processed/'+ category +'/'+country +'/'+ state +'/'+ city+'/'+ js)
            #the hierarchy of folder structures is like: Data Processsed/Category/Country/State/City/.json
            
```

# Question 2_Part2
1. Read all the json files in the folder called Data.
2. Read only the json files which contain the key called “restaurants”
3. Each (or most of the json files) contain a key called “open” which contains the details of the operation (timings) of the restaurants. For each json file, read the timings of the restaurants.
4. Data of the operation timings of the restaurants is present for each day of the week. I want you to extract each of this data and write it in an excel sheet.
5. Example:
6. For a particular restaurant named “The Coffee Grounds”, the excel sheet should look like this:

```python
import os, json
import csv
```

```python
#get all json files in folder Data.
path_to_json='Data/'
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]

# step: find all json the type is restaurant and put them into one folder
for js in json_files:
    #define the list/column and the title of each column
    nameList= []
    cityList=[]
    CountryList=[]
    starList=[]
    endList=[]
    dayList=[]    
    dayList.insert(0,'Day of Week')
    starList.insert(0,'Start Time')
    endList.insert(0,'End Time')
    
    # open and read each json file in the Data folder.
    with open(os.path.join(path_to_json, js)) as json_file:     
        #convert JSON to python List Object
        alldata= json.load(json_file)
        # get the json file that term is 'restaurant'
        category = alldata['term']
        if category=='restaurants':
            #deal with exception for example: json file contains error value or error key or error type
            try:
                #get the data of the country,city,state,name,openday,start time,and end time.
                country = alldata['location']['country']    
                city = alldata['location']['city']
                state = alldata['location']['state']        
                name = alldata['name']
#                 res_id = alldata['id']
                openday = alldata['hours']
                for day in openday:             
                    for one in day['open']: 
                        dayList.append(one['day'])
                        starList.append(one['start'])                
                        endList.append(one['end'])
                
                #get the list of name according to the number of openday
                nameList= [name+'',] * (len(dayList)-1)
                nameList.insert(0,'Name of Restaurant')
                
                #get the list of city according to the number of openday
                cityList =[city+'',] * (len(dayList)-1)
                cityList.insert(0,'City')
                
                #get the list of country according to the number of openday                
                CountryList=[country+'',]*(len(dayList)-1)     
                CountryList.insert(0,'Country Code')
                 
                #write all the list to each column of the csv named 'restaurantsAll.csv' and use 'a' to append every loop of rows
                #csv_out = open ('res/'+ res_id+'.csv','w')
                csv_out = open ('restaurantsAll.csv','a')
                mywriter = csv. writer(csv_out)            
                for row in zip(nameList,cityList,CountryList,dayList,starList,endList):
                    mywriter.writerow(row)              
                csv_out.close()
                
            except (ValueError, KeyError, TypeError):
                print ("JSON format error" )   
 
```

* Bonus: Split “Start Time” column into two different columns having the hour and minute in each of them
* Split “End Time” column into two different columns having the hour and minute in each of them

```python
#get all json files in folder Data.
path_to_json='Data/'
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]

# step: find all json the type is restaurant and put them into one folder
for js in json_files:
    #define the list/column and the title of each column
    nameList= []
    cityList=[]
    CountryList=[]
    starMinList=[]
    startHList=[]
    endHList=[]
    endMinList=[]
    dayList=[]
    
    dayList.insert(0,'Day of Week')
    startHList.insert(0,'Start Time Hour')
    starMinList.insert(0,'Start Time Minute')
    endHList.insert(0,'End Time Hour')
    endMinList.insert(0,'End Time Minute')
    
    # open and read each json file in the Data folder.
    with open(os.path.join(path_to_json, js)) as json_file:     
        #convert JSON to python List Object
        alldata= json.load(json_file)
        # get the json file that term is 'restaurant'
        category = alldata['term']
        if category=='restaurants':
            #deal with exception for example: json file contains error value or error key or error type
            try:
                #get the data of the country,city,state,name,openday,start time,and end time.
                country = alldata['location']['country']    
                city = alldata['location']['city']
                state = alldata['location']['state']        
                name = alldata['name']
#                 res_id = alldata['id']
                openday = alldata['hours']
                for day in openday:             
                    for one in day['open']: 
                        dayList.append(one['day'])
                        
                        #there is a if statement to splite the start and end time
                        starMinList.append(one['start'][-2:])                 
                        if(len(one['start']))==3:   
                            startHList.append(one['start'][:1])
                        else:
                            startHList.append(one['start'][:2])
                    
                        endMinList.append(one['end'][-2:])                 
                        if(len(one['end']))==3:   
                            endHList.append(one['end'][:1])
                        else:
                            endHList.append(one['end'][:2])
                
                #get the list of name according to the number of openday
                nameList= [name+'',] * (len(dayList)-1)
                nameList.insert(0,'Name of Restaurant')
                
                #get the list of city according to the number of openday
                cityList =[city+'',] * (len(dayList)-1)
                cityList.insert(0,'City')
                
                #get the list of country according to the number of openday                
                CountryList=[country+'',]*(len(dayList)-1)     
                CountryList.insert(0,'Country Code')
                 
                #write all the list to each column of the csv named 'restaurantsAll2.csv' and use 'a' to append every loop of rows
                #csv_out = open ('res/'+ res_id+'.csv','w')
                csv_out = open ('restaurantsAll2.csv','a')
                mywriter = csv. writer(csv_out)            
                for row in zip(nameList,cityList,CountryList,dayList,startHList,starMinList,endHList,endMinList):         

                    mywriter.writerow(row)              
                csv_out.close()
                
            except (ValueError, KeyError, TypeError):
                print ("JSON format error" )   
 
```
